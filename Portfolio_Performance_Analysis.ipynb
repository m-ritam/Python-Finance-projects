```python
file_path = "/content/drive/My Drive/Excel file to use in colab/4_historical_stock_prices.xlsx"
```

```python
df4 = pd4.read_excel(file_path)
```

```python
returns_df4 = df4.set_index("Date").pct_change() * 100  # creating 'Date' as index & Convert to percentage

returns_df4.reset_index(inplace=True)   # Returning 'Date' back to a regular column
```

```python
returns_df4
```

```python
cumulative_returns = (1 + returns_df4.set_index("Date") / 100).cumprod() - 1
cumulative_returns.reset_index(inplace=True)
```

```python
cumulative_returns
```

```python
volatility = returns_df4.set_index("Date").std()
```
